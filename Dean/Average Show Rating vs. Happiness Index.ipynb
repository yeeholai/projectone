```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
netflix_country = pd.read_csv("Netflix_data/netflix_country.csv")
happiness = pd.read_csv("Happiness/2021.csv")
```

```python
netflix_country.head(3)
```

```python
happiness.head(3)
happiness_mergeready = happiness[['Country name', 'Ladder score']]
happiness_mergeready
```

```python
netflix_countrymean = netflix_country.groupby(by="Country Availability")[['Hidden Gem Score', 'IMDb Score', 'Rotten Tomatoes Score', 'Metacritic Score']].mean()
```

```python
netflix_countrymean.reset_index(inplace=True)
netflix_countrymean
```

```python
netflix_countrymean.rename(columns={"Country Availability":"Country name"},inplace=True)
netflix_countrymean.head(2)
```

```python
happiness_mergeready
```

```python
happynetflix = pd.merge(netflix_countrymean, happiness_mergeready,how='inner', on='Country name')
happynetflix
```

```python
happynetflix.rename(columns={"Ladder score":"Happiness Index"},inplace=True)
```

```python
happynetflix.head(2)
```

```python
slope, intercept, r, p, se = st.linregress(happynetflix['Happiness Index'],happynetflix['Hidden Gem Score'])
line_eq= "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(happynetflix['Happiness Index'], happynetflix['Hidden Gem Score'])
plt.xlabel("Happiness Index")
plt.ylabel("Hidden Gem Score")
plt.title("Happiness Index vs. Hidden Gem Score")
plt.plot(happynetflix['Happiness Index'], intercept+slope*happynetflix['Happiness Index'],'r',label='fitted line')
plt.annotate(line_eq,(4,6.2),fontsize=15,color='red')
plt.savefig("Output/happyvhiddengem.png")
plt.show()
print(f"The r-value is {round(r,4)}, showing a slight negative correlation between Happiness Index and the Hidden Gem Score")
```
